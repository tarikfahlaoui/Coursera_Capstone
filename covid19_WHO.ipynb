```python
# A modifier

def search_table(list_df):
    for index in range(len(list_df)):
        for column in list_df[index].columns:
            if np.sum(list_df[index][column].str.contains('Total'))>0:
                df = list_df[index]
    return df
```

```python
# A modifier

def clean_table(df):
    df = df.iloc[2:-2,1:].rename(columns={1:'Country',2:'Confirmed_cases'}).set_index('Country')
    df['Confirmed_cases'] = df['Confirmed_cases'].astype(int)
    return df
```
